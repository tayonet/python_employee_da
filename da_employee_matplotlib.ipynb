```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python


employees = pd.read_csv("analysed_employees.csv")
employees.head()
```

```python
# show a bar chart of the number of employees in each department
employees["dept"].value_counts().plot(kind="bar" , title="Employee Count By Dept")
plt.show()
```

```python
# show the percentage distribution by salary_levels using the pie chart
employees["salary_levels"].value_counts().plot(kind="pie", autopct="%1.1f%%",title="Salary levels")
plt.show()
```

```python
# show the pie chart distribution of employees for each departments
employees["dept"].value_counts().plot(kind="pie", autopct="%1.1f%%",title="Dept %")
plt.show()
```

```python
# show the percentage distribution by gender using the pie chart
employees["gender"].value_counts().plot(kind="pie", autopct="%1.1f%%",title="Gender %")
plt.show()
```

```python
# show the histogram employees salaries 
employees["salary"].plot(kind="hist", title = "Histogram of Salary", bins = 10)
plt.grid()
plt.show()
```

```python
# show the summary statistics using boxplot for salaries by departments
employees.boxplot(column= "salary", by = "dept")
plt.suptitle("Summary stats by dept")
plt.show()
```

```python
# show the summary statistics of salaries based on the employees gender
employees.boxplot(column="salary",by="gender")
plt.suptitle("Summary stats by gender")
plt.show()
```

```python
# Show the line chart for the number of employees joined per year
employees.groupby("hire_year")["hire_year"].value_counts().plot(kind="line", marker= "x",title="Line chart")
plt.xlabel("Years")
plt.ylabel("Count of Employee")
plt.grid()
plt.show()
```

```python
# horizontal bar chart of average salary by department
employees.groupby("dept")["salary"].mean().sort_values(ascending=True).plot(kind="barh" , title= "Avg by dept")
plt.show()

```

```python

# stacked bar chart of gender count by department
pd.crosstab(employees["dept"],employees["gender"]).plot(kind="bar" , stacked=True,title="Count by dept n gender")
plt.show()
```

```python
# show the correlation between salary and age using the scatter plot
employees.plot(kind="scatter" , x = "age" , y= "salary", title ="Correlation btw age and salary")
plt.show()
```

```python
# show correlation between salary and years of service
employees.plot(kind="scatter", x = "yos" , y = "salary", title= "Correlation btw yos and salary")
plt.show()
```

```python
# show highest 3 salary earners in the dataset using bar chart
employees.nlargest(3,"salary").set_index("fullname")["salary"].plot(kind= "bar", title="Highest 3 salary earners")
plt.show()
```
